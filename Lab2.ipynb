# **# Using Pandas**

```python
import numpy as np
import pandas as pd
```

```python
person= pd.read_csv('cars.csv')
```

#1).ScatterPlot

```python
person.plot.scatter(x='age',y='income',title='Age Vs Income',color='purple',figsize=(10,3),s=50,alpha=0.9)

# As we can see in below garph that age of person increasing with Income which means that 
# as Age of person increseing with the Income of him/her
```

#2).Line

```python
person.head(20).plot.line(y='income',title='Income',color='red')

# Income of person is increasing trend with short period of time
```

#3).Histrogram

```python
person.plot.hist(y='age')

# Highest frequency of age is between 25 to 28 and Lowest frequency of age is between 50 to 55
# It means the younger people are more than the aged people
```

#4).(i)Bar

```python
person['income'].value_counts().sort_index().plot.bar()

# Most of the person has higher income than the other of them
# and there are lot of people which lay in each range of Income 
```

#(ii).Barh

```python
person['age'].value_counts().sort_index().plot.barh()

# Most of the person has higher income than the other them
# and there are lot of people which lie in each range of Income 
```

#5).Boxplot

```python
person['income'].plot.box()

#In the below graph the medain value is 80000 and most of people income lie in lower quartile
```

# 6).Densityplot

```python
person['income'].plot.kde(x='Income',title='Density of Income')

# As we can see from graph that most of the people having income between 50000 to 100000
# and there is no skew which mean that mean and median are same
```

# 7).Hexbin plot

```python
person.plot.hexbin(x='age',y='income',gridsize=20)

# In the below graph most peoples income lie in the range 40000-50000
# which is why the hexagonal bin is darker compare to other bins 
```

# 8).Pie chart

```python
person.groupby(['gender']).sum().plot.pie(y='age', autopct='%1.0f%%')

# we can see in below graph that male and female are have same frequency 
```

**# Using Matplotlib**

```python
import matplotlib.pyplot as plt
```

1).(i)ScatterPlot

```python
fig,ax=plt.subplots()
ax.scatter(person['age'],person['income'])
ax.set_title('Age Vs. Income')
ax.set_xlabel('Age')
ax.set_ylabel('income')

# As we can see in below garph that age of person increasing with Income which means that 
# as Age of person increseing with the Income of him/her
```

(ii).Scatterplot

2).Line Plot

```python
plt.plot(person['Age'],color='red',linestyle='--',linewidth='1')
plt.plot(person['Income'])
plt.show()

#Income of person is increasing trend with age as shown in graph
```

3).Histogram plot

```python
plt.hist(person['Age'])
plt.title('Frequency distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Highest frequency of age is between 25 to 28 and Lowest frequency of age is between 50 to 55
# It means the younger people are more than the aged people
```

4).Bar plot

```python
data =person['Gender'].value_counts()
x=data.index
y=data.values
plt.bar(x,y)
plt.title('Count of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')

# Number of male greater that number of Female as shown in graph
```

5).Pie chart

```python
data =person['Gender'].value_counts()
x=data.index
y=data.values

plt.pie(y,labels=x)

# we can see in below graph that male and female are have same frequency 
```

6).Box plot

```python
person.plot.box(['Age','income'])

#In the below graph the medain value is 80000 and most of people income lie in lower quartile
```

7).hexbin

```python
plt.hexbin(x=person['Age'], y=person['Income'], gridsize=15, bins=10, xscale='linear', yscale='linear', marginals=False)

# In the below graph most peoples income lie in the range 60000-80000 and 20000-40000
# in this incomes are represented by different color of hexagonal bin
```
